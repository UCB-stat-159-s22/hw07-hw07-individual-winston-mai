Claim : Because of global warming, wildfires happen more and more frequently, which means as time goes, we will see more wildfires and potentially more major ones.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
#from datetime import datetime
```

```python
data = pd.read_csv("data/California_Fire_Incidents.csv")
data = data[['ArchiveYear', 'AcresBurned', 'Counties', 'Started', 'Extinguished', 'MajorIncident']]
```

```python
data['Started'] = pd.to_datetime(data['Started'])
data['Extinguished'] = pd.to_datetime(data['Extinguished'])
data.fillna(0)
data['AcresBurned'] = data['AcresBurned'].astype('float64')

data
```

```python
plt.figsize = (12,7)
data['ArchiveYear'].value_counts().sort_index().plot(kind='bar', figsize=(12,7))

plt.xlabel('Year')
plt.ylabel('Number of Fire')
plt.title('Frequency of Fire per Year in California (2013-2019)')
plt.savefig("figures/Fire_Count.png");
```

```python
data['Counties'].value_counts().sort_index().plot(kind='bar', figsize=(12,7))

plt.xlabel('County Name')
plt.ylabel('Number of Fire')
plt.title('Counties (in California) where Fires Occur Between 2013-2019')
plt.savefig("figures/NumberByCounty.png");
```

```python
data[['Counties', 'AcresBurned']].groupby('Counties').sum().plot(kind='bar', figsize=(12,7)) 

plt.xlabel('County Name')
plt.ylabel('Acres Burned')
plt.title('Acres Burned by County Between 2013-2019')
plt.savefig("figures/Acres_Burned.png");
```
